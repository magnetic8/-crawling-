```python
# 코드를 실행시키지 않는 시간을 부여
import time
```

```python
# 상품 10개에 대한 정보를 수집
driver = wb.Chrome()
driver.get('http://corners.gmarket.co.kr/Bestsellers')
title_list=[]
price_list=[]
for i in range(200):   
    # 200개의 이미지 찾기
    imgs = driver.find_elements(By.CLASS_NAME , 'lazy')
    # i번째 클릭
    imgs[i].click()
    # 1초동안 코드가 실행이 안됨
    time.sleep(1)
    # 상품명 가져오기
    title = driver.find_element(By.CLASS_NAME, 'itemtit')
    # 가격 가져오기
    price = driver.find_element(By.CLASS_NAME, 'price_real')
    title_list.append(title.text)
    price_list.append(price.text)
    # 뒤로 가기
    print(f'상품명 :{title.text} ')
    print(f'가격 :{price.text} ')
    print(i)
    driver.back()
    # 1초동안 코드가 실행이 안됨
    time.sleep(1)
```

```python
import pandas as pd
# 딕셔너리 만들기
dic = {'상품명':title_list, '가격': price_list}
# 데이터프레임 만들기
df = pd.DataFrame(dic)
# 엑셀 파일로 만들기
df.to_csv('gmarket_best.csv', encoding='euc-kr')
```
