```python
# 멜론을 속이고 크롬인척 성공!
res = req.get('https://www.melon.com/chart/index.htm',headers = user_agent)
```

```python
soup=bs(res.text,'lxml')
```

### 1. 노래 제목 수집

```python
# > 자식, ' ' 자손
title = soup.select('div.ellipsis.rank01 > span > a')
# data = soup.select('div  a') 자손
```

```python
title[0].text
```

```python
for i in title:
    print(i.text)
```

### 2. 가수 수집

```python
# singer = soup.select('div.ellipsis.rank02 > span > a')
# a태그는 하나의 노래에 여러가수가 있을 때 실패
# 가수 각각 a태그에 담겨져 있기 때문
```

```python
# 가수 전체가 포함된 태그를 수집
# div 태그로 가져오면 안에 같은 값이 두 번 나옴
#singer = soup.select('div.ellipsis.rank02')
```

```python
# span 태그를 활용해서 가수 수집
singer = soup.select('span.checkEllipsis')
len(singer)
```

```python
for i in singer:
    print(i.text)
```

```python
# 0726
# 수집한 데이터를 사용자가 사용할 수 있게 변화 > 엑셀파일로 만들기
```

```python
# 리스트를 만들어서 title과 singer의 값을 저장
title_list =[]
singer_list = []
for i in range(len(singer)):
    singer_list.append(singer[i].text)
    title_list.append(title[i].text)
    
```

```python
title_list
```

```python
# 리스트로 만든 데이터를 표(Data Frame)형식으로 제작
import pandas as pd
# 리스트로 데이터프레임을 만들면 행 방향으로 데이터 프레임이 만들어짐
df1 = pd.DataFrame([singer_list,title_list])
```

```python
# 전치 함수 사용하기
df1 = df1.T
```

```python
# 컬럼 이름 변경
df1.columns = ['가수','노래']
df1
```

```python
# 딕셔너리로 데이터프레임 만들기
dic = {'가수':singer_list, '노래':title_list}
pd.DataFrame(dic)
```

```python
# 인덱스 번호 변경
df1.index = range(1,101)
df1
```

```python
# 인덱스에 이름 달기
df1.index.name = '순위'
df1
```

```python
# csv파일로 만들기
df1.to_csv('melon_tip100.csv', encoding = 'euc-kr')
```
