```python
from selenium import webdriver as wb
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
import time
```

```python
driver =wb.Chrome()
driver.get('https://map.naver.com/v5/search/%EA%B4%91%EC%A3%BC%EB%A7%9B%EC%A7%91?c=14117032.5098384,4185726.6182519,11,0,0,0,dh')
```

```python
# id="searchIframe" 에 접근
driver.switch_to.frame('searchIframe')
```

```python
# iframe 안에 있기 때문에 지금은 오류가 남
# iframe 안에 접근해야 함
# iframe이 가지고 있는 src 확인 > about:blank = 비어 있음
# switch_frame() > 태그 정보로 iframe 에 접근

title = driver.find_elements(By.CLASS_NAME, 'place_bluelink.OXiLu')
for i in title:
    print(i.text)
```

```python
# iframe 에서 빠져나오기
driver.switch_to.default_content()
```

```python
# 음식점 button태그 
# 클래스 bubble_filter ng-star-inserted
driver.find_element(By.CSS_SELECTOR, 'button.bubble_filter.ng-star-inserted').text
```
