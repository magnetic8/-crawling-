```python
# 100등까지의 종목명, 현재가, 거래량 수집
# csv파일로 저장
import requests as req
from bs4 import BeautifulSoup as bs
```

```python
# 사이트 접근하기
res = req.get('https://finance.naver.com/sise/sise_quant.naver?sosok=0')
# html로 변경하기
soup = bs(res.text, 'lxml')
```

```python
# 종목명 받아오기
title = soup.select('a.tltle')
title[0].text
```

```python
len(title)
```

```python
title_list = []
for i in range(len(title)):
    title_list.append(title[i].text)
```

```python
# copy selector 를 사용하면 가지고 와지는 태그와 선택자는 
# 해당하는 대상 하나를 기준으로 하기 때문에
# 원하는 여러 대상을 가져올경우에는 수정을 반드시 해줘야 함
```

```python
# 현재가 받아오기
# nth-child() : 똑같은 태그가 여러개 있으면 순서대로 지정
price = soup.select('td.number:nth-child(3)')
price_list=[]
for i in range(len(price)):
     price_list.append(price[i].text)
```

```python
# 오른쪽 마우스 클릭- copy-copy select
#contentarea > div.box_type_l > table > tbody > tr:nth-child(3) > td:nth-child(3)
```

```python
# 거래량 받아오기
trade = soup.select('td.number:nth-child(6)')
trade_list =[]
for i in range(len(trade)):
     trade_list.append(trade[i].text)
```

```python
# 딕셔너리로 합쳐주기
dic = {'종목명':title_list, '현재가':price_list, '거래량':trade_list}
```

```python
#  데이터프레임으로 변환하기
import pandas as pd
df = pd.DataFrame(dic)
df
```

```python
# 엑셀 파일 만들기
df.to_csv('220726주가.csv', encoding = 'euc-kr')
```
