```python
import requests as req
from bs4 import BeautifulSoup as bs
```

```python
# 1. req를 통해서 리뷰사이트 데이터 요청
# https://movie.naver.com/movie/bi/mi/point.naver?code=81888
res = req.get('https://movie.naver.com/movie/bi/mi/point.naver?code=81888')
```

```python
# 2. 컴퓨터가 이해할 수 있는 언어(HTML)로 변경
soup = bs(res.text, 'lxml')
```

```python
# 3. soup데이터에서 select를 통해 리뷰 정보만 수집
# 눈에 보이지만 수집할 수 없음
# 해당 태그는 iframe 태그로 만들어져 있음 
soup.select('div.score_reple>p')
```

### iframe의 개념
- iframe은 창을 통해서 다른 서버(주소)에 있는 정보를 보여줄 때 사용
- 1. f12 개발자 도구(ctrl + f) iframe 검색
- 2. 찾고자 하는 데이터가 iframe에 속해있는지 확인(화면이 파란색으로 변경)
- 3. 진짜 주소는 iframe의 src속성에 있음 
- 4. src속성에 있는 주소는 메인 서버주소가 빠져있음.
    -  iframe 에 주소가 없는 경우가 가끔 있음 > 수집불가

```python
res = req.get('https://movie.naver.com/movie/bi/mi/pointWriteFormList.naver?code=81888&type=after&i[…]geSubscriptionAlready=false&isMileageSubscriptionReject=false')
soup = bs(res.text,'lxml')
```

```python
# span태그에 두 개의 값이 나옴
# 관람객 > 1번
# 리뷰 > 2번
review = soup.select('div.score_reple > p > span:nth-child(2)')
```

```python
# 불필요한 글자가 같이 수집
# 띄어쓰기, 엔터, 탭 > strip()
# strip() 앞과 뒤의 띄어쓰기 전부 삭제
review[0].text.strip()
```

```python
# 12페이지에는 관람객이 아닌 리뷰도 존재
# 불필요한 관람객이라는 글자들만 삭제(추출)
```

```python
ico = soup.select('span.ico_viewer')

# 반복문을 사용해서 하나씩 삭제(추출)
# extract() > soup데이터에서 해당하는 태그를 삭제
for i in ico:
    i.extract() #관람객이 삭제
```

# 1페이지부터 10페이지까지 수집 

```python
# 1페이지 수집을 for문 돌릴 수 있게 모아두기
```

```python
res = req.get('https://movie.naver.com/movie/bi/mi/pointWriteFormList.naver?code=81888&type=after&i[…]riptionAlready=false&isMileageSubscriptionReject=false&page=1')
soup = bs(res.text,'lxml')
ico = soup.select('span.ico_viewer') # 관람객 태그
for i in ico :
    i.extract() # 관람객 삭제 
review = soup.select('div.score_reple > p > span')
for i in review:
    print(i.text.strip())
```

```python
# 1~10페이지까지 확장
for page in tqdm_notebook (range(1,101)):
    res = req.get(f'https://movie.naver.com/movie/bi/mi/pointWriteFormList.naver?code=81888&type=after&i[…]riptionAlready=false&isMileageSubscriptionReject=false&page={page}')
    soup = bs(res.text,'lxml')
    ico = soup.select('span.ico_viewer') # 관람객 태그
    for i in ico :
        i.extract() # 관람객 삭제 
    review = soup.select('div.score_reple > p > span')
    for i in review:
        print(i.text.strip())
```
