```python
from selenium import webdriver as wb
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
import time
```

```python
# 1. 브라우저 실행
driver = wb.Chrome()
driver.get('https://www.youtube.com/watch?v=3ERtNZqh1XA')
```

```python
# 2. 댓글정보를 수집
## 스크롤이 바닥을 찍으면 20개의 댓글이 로딩
review = driver.find_elements(By.ID,'content-text')
len(review)
```

```python
for i in review:
    print(i.text)
```

```python
# 3. 댓글 정보를 갱신하기 위해서 키보드의 end 사용하기 
## body 태그에 end 버튼을 눌러줌
while True:
    
    body = driver.find_element(By.TAG_NAME, 'body')
    body.send_keys(Keys.END)
    time.sleep(0.5)
```

```python
for i in range(100):
    body = driver.find_element(By.TAG_NAME, 'body')
    body.send_keys(Keys.END)
    time.sleep(0.5)   
```
